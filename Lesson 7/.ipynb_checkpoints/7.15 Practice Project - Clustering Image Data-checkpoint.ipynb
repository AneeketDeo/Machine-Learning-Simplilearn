```python
# Objective: To identify the dominant color in the image.

# Action to Perform:

# 1.Open and display the image “dog.jpeg.”
# 2.Use K-means clustering for image segmentation, which will include the following steps:
# 3.Find out the dimensions of the image and convert it to a two-dimensional array.
# 4.Use K-means clustering with k set to 3 and cluster the image.
# [Hint: Refer to the K-means module of scikit learn]
# 5.Predict the cluster label of every pixel in the image and plot it back as an image.
# 6.Find out the three dominant colors in the image.
# [Hint: The cluster centers should correspond to three dominant colors]
```

```python
# IMPORTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image as pil
```

# USING PIL

```python
# IMAGE

# 1.Open and display the image “dog.jpeg.”

img = pil.open('dogs.jpeg')
img
```

```python
# 2.Use K-means clustering for image segmentation, which will include the following steps:

# CONVERT IMG INTO NUMPY ARRAY
npimg = np.asarray(img)
npimg
```

```python
# 3.Find out the dimensions of the image and convert it to a two-dimensional array.

# DIMENSIONS OF IMAGE
print(npimg.shape)
```

```python
# CONVERT TO TWO DIM
new_img = npimg.reshape(185 * 272, 3)
```

```python
# 4.Use K-means clustering with k set to 3 and cluster the image.
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(new_img)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
# 5.Predict the cluster label of every pixel in the image and plot it back as an image.

label = model.predict(new_img)
label = label.reshape((185, 272))
plt.imshow(label)
```

```python
plt.imshow(model.cluster_centers_)
```

```python
plt.imshow(model.cluster_centers_[0]) # gives invalid shape --> THATS WHY WE NEED AN EMPTY ARRAY TO CONVERT 
# THIS ARRAY INTO 3 DIM
```

```python
# CORRECT WAY
emp_arr = np.ones((100, 100, 3)) #creates an empty array with 'ones'

centreone = emp_arr * model.cluster_centers_[0] #it miscalculates here... so we need to traverse the array separately

plt.imshow(centre1)
```

```python
# FINAL CORRECT WAY
centers = [centers.astype(int) for centers in model.cluster_centers_]
centre1 = emp_arr * centers[0]
plt.imshow(centre1)
```

```python
centre2 = emp_arr * centers[1]
plt.imshow(centre2)
```

```python
centre3 = emp_arr * centers[2]
plt.imshow(centre3)
```
